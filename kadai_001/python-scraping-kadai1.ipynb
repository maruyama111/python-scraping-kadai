```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://news.yahoo.co.jp/articles/6c8aa4e0e386e15c24315df06c209bb5031bfbfe'
```

```python
response= requests.get(url)
```

```python
soup=BeautifulSoup(response.text, 'html.parser')
```

```python
title_element = soup.select_one('#uamods >  .article_body > div> p')
```

```python
title_text = title_element.text
print(title_text)
```
