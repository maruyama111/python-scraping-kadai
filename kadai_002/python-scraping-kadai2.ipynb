```python
import requests
import pprint
from getpass import getpass
```

```python
api_key = getpass('APIキーを入力してください')
```

```python
station_name = input('駅名を入力してください')
```

```python
# 駅情報　APIエンドポイントURL（APIアクセス接続先）
url = 'https://maps.googleapis.com/maps/api/geocode/json'

# データを取得
response = requests.get(
    url,
    params={
        'key':api_key,
        'address':station_name
    }
)
```

```python
# レスポンスのJSONを取得
json_data = response.json()

# JSONの中身を確認
pprint.pprint(json_data)
```

```python
for item in json_data['results']:
  map_info = item['address_components'][0]['short_name']
  location_lat=item['geometry']['location']['lat']
  location_lng=item['geometry']['location']['lng']
```

```python
# レストラン情報　APIエンドポイントURL（APIアクセス接続先）
url2 = f'https://maps.googleapis.com/maps/api/place/nearbysearch/json?keyword={station_name}&location={location_lat}%2C{location_lng}&radius=500&language=japan&type=restaurant&key={api_key}'

# データを取得
response2 = requests.get(url2)
json_data2 = response2.json()

pprint.pprint(json_data2)
```

```python
for results in json_data2['results']:
  print(f"{results['name']} : {results['rating']} : {results['vicinity']}")
```
